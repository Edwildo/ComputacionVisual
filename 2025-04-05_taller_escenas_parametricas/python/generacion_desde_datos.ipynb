```python
from vedo import Cube, Sphere, Plotter, write
import numpy as np
import random
```

```python
# Coordenadas manuales
coords = np.array([
    [0, 0, 0],
    [2, 0, 0],
    [4, 0, 0],
    [0, 2, 0],
    [2, 2, 0],
    [4, 2, 0],
])
```

```python
# Crear objetos 3D desde coordenadas
objects = []

for i, (x, y, z) in enumerate(coords):
    # Alternar entre esfera y cubo
    if i % 2 == 0:
        shape = Sphere(pos=(x, y, z), r=0.6)
        shape.color('skyblue')
    else:
        shape = Cube(pos=(x, y, z))  # ✅ tamaño por defecto
        shape.color('orange')
    
    # Escalar en función del índice
    shape.scale(1 + i * 0.1)
    objects.append(shape)
```

```python
# Mostrar los objetos generados
plotter = Plotter(title="Escena Paramétrica", axes=1)
plotter.show(objects)
```

```python
from vedo import merge

# Unir todos los objetos en una sola malla para exportar bien
merged_scene = merge(objects)

# Exportar a .OBJ
write(merged_scene, "scene_parametrica.obj")

# También puedes descomentar una de estas líneas si deseas exportar a otro formato:
# write(merged_scene, "scene_parametrica.stl")
# write(merged_scene, "scene_parametrica.gltf")

```
